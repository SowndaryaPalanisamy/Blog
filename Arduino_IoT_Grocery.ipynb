```
#include <ESP8266WiFi.h>;
 
#include <WiFiClient.h>;
 
#include <ThingSpeak.h>;

#include "DHT.h"        // including the library of DHT11 temperature and humidity sensor
#define DHTTYPE DHT11   // DHT 11

#define dht_dpin D4
DHT dht(dht_dpin, DHTTYPE); 

const int trigP = D1;  //D4 Or GPIO-2 of nodemcu
const int echoP = D2;  //D3 Or GPIO-0 of nodemcu

const int mq3 = D3;

//the analog pin of the alcohol sensor goes into analog pin A0 of the arduino

int limit;
int value;

long duration;
int distance;

const char* ssid = "DHT"; //Your Network SSID
 
const char* password = "12345678"; //Your Network Password
 
WiFiClient client;
 
unsigned long myChannelNumber = 1079726; //Your Channel Number (Without Brackets)
 
const char * myWriteAPIKey = "KS8DMH5FCA22OP2O"; //Your Write API Key
 
void setup()
 
{
  dht.begin();
  pinMode(trigP, OUTPUT);  // Sets the trigPin as an Output
pinMode(echoP, INPUT);   // Sets the echoPin as an Input
Serial.begin(115200);
  Serial.println("Humidity and temperature\n\n"); 
delay(10);
 
// Connect to WiFi network
 
WiFi.begin(ssid, password);
 
 
 
ThingSpeak.begin(client);
 
}
 
 
 
void loop()
 
{
digitalWrite(trigP, LOW);   // Makes trigPin low
delayMicroseconds(2);       // 2 micro second delay 

digitalWrite(trigP, HIGH);  // tigPin high
delayMicroseconds(10);      // trigPin high for 10 micro seconds
digitalWrite(trigP, LOW);   // trigPin low

duration = pulseIn(echoP, HIGH);   //Read echo pin, time in microseconds
distance= duration*0.034/2;        //Calculating actual/real distance

Serial.print("Distance = ");        //Output distance on arduino serial monitor 
Serial.println(distance);

 value = digitalRead(mq3); //reads the analaog value from the alcohol sensor's AOUT pin
  Serial.print("Alcohol value: ");
  Serial.println(value);//prints the alcohol value
 
float h = dht.readHumidity();
    float t = dht.readTemperature();         
    Serial.print("Current humidity = ");
    Serial.print(h);
    Serial.print("%  ");
    Serial.print("temperature = ");
    Serial.print(t); 
    Serial.println("C  ");
    delay(800);
 
ThingSpeak.writeField(myChannelNumber, 1,t, myWriteAPIKey); //Update in ThingSpeak
ThingSpeak.writeField(myChannelNumber, 2,h, myWriteAPIKey); //Update in ThingSpeak
ThingSpeak.writeField(myChannelNumber, 3,distance, myWriteAPIKey); //Update in ThingSpeak 
ThingSpeak.writeField(myChannelNumber, 4,value, myWriteAPIKey); 
 
delay(100);
 
}
```
